```python
import pandas as pd
reviews = pd.read_csv("./winemag-data-130k-v2.csv", index_col=0)
pd.set_option('display.max_rows', 5)
```

```python
reviews
```

```python
reviews.country
reviews['country']
```

```python
# Index-based selection
reviews.iloc[1]
```

```python
reviews.iloc[:, 0]
```

```python
# select just the second and third entries
reviews.iloc[1:3, :2]
```

```python
reviews.iloc[[1,2,4,35,410], [2,4,-1]]
```

```python
# Label-based selection
reviews.loc[:, 'price']
```

```python
reviews.loc[123:, ['country', 'variety', 'province', 'region_1', 'region_2']]
```

### iloc uses the Python stdlib indexing scheme, where the first element of the range is included and the last one excluded. So 0:10 will select entries 0,...,9. loc, meanwhile, indexes inclusively. So 0:10 will select entries 0,...,10.

```python
# Manipulating the index
reviews.set_index("country")
```

```python
# Conditional Selecting
reviews.country == 'Italy'
```

```python
reviews.loc[reviews.country == 'Italy']
```

```python
reviews.loc[(reviews.country=='US') & (reviews.price >= 100)]
```

```python
reviews.loc[(reviews.country=='None') | (reviews.price >= 100)]
```

```python
reviews.loc[reviews.country.isin(['US', 'France'])]
```

```python
reviews.loc[reviews.price.notnull()]
```

```python
# Assigning data
reviews['misc'] = True
reviews.iloc[:, -1:] 
```

```python
reviews['price2'] = reviews['price'].multiply(2)
reviews.loc[reviews.price.notnull(), ['price','price2']]
```
